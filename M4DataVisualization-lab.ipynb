## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)

```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# your code goes here

QUERY = """
SELECT ConvertedComp
FROM  master
"""

df=pd.read_sql_query(QUERY,conn)

df['ConvertedComp'].dropna().plot(kind='hist', bins=40, edgecolor='white', color='skyblue')
plt.xlabel('Compensation')
plt.title('Compensation Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age 
FROM  master
"""

df=pd.read_sql_query(QUERY,conn)
df['Age'].dropna().plot(kind='box',color='skyblue')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM  master
"""

df=pd.read_sql_query(QUERY,conn)

#df.columns
df.plot(kind='scatter',x='Age', y='WorkWeekHrs', color='skyblue', s=50, alpha=0.7)
```

### Bubble Plots


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs,CodeRevHrs
FROM  master
"""

df=pd.read_sql_query(QUERY,conn)


# Create bubble plot
df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s=df['Age']*2, color='skyblue', alpha=0.8, label='Age')

plt.title('Bubble Plot of CodeRevHrs vs. WorkWeekHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend()
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
PRAGMA table_info(master)
"""

df_columns = pd.read_sql_query(QUERY, conn)
print(df_columns)
```

```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as NumberLearners
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY NumberLearners DESC
LIMIT 5
"""

df=pd.read_sql_query(QUERY,conn)
print(df.head(5))

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
df.set_index('DatabaseDesireNextYear')['NumberLearners'].plot(
    kind='pie', 
    figsize=(8, 8), 
    autopct='%1.0f%%',
    colors=colors, 
    legend=False
)

plt.title('Top 5 Most Desired DBs')
plt.legend(labels=df['NumberLearners'], loc='upper left', bbox_to_anchor=(1, 0.8))
plt.ylabel('')  
plt.show()
```

```python
# your code goes here

QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as NumberLearners
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY NumberLearners DESC
LIMIT 5
"""

df=pd.read_sql_query(QUERY,conn)
print(df.head(5))

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
df.set_index('LanguageDesireNextYear')['NumberLearners'].plot(
    kind='pie', 
    figsize=(8, 8), 
    autopct='%1.0f%%',
    colors=colors, 
    legend=False
)

plt.title('Top 5 Most Desired Languages')
plt.legend(labels=df['NumberLearners'], loc='upper left', bbox_to_anchor=(1, 0.8))
plt.ylabel('')  
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df_wrkwk = pd.read_sql_query(QUERY, conn)
df_wrkwk.head()
colors = ['#ff9999', '#c2c2f0']
df_graph = df_wrkwk.groupby("Age").median()
df_graph.plot(kind='bar', stacked=True, color=colors)
plt.title('Median workweek hours and CodeRevHrs in the 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_compage = pd.read_sql_query(QUERY, conn)
df_compage.head()
df_line = df_compage.groupby("Age").median()
df_line['ConvertedComp'].plot(kind='line', color='skyblue')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master 
"""
df_branch = pd.read_sql_query(QUERY, conn)
df_hbar = df_branch['MainBranch'].value_counts()
colors = ['#ff9999', '#c2c2f0']
df_hbar.plot(kind='barh', color=colors)
plt.title('Frequency of Responses')
plt.xlabel('Frequency')
plt.ylabel('Main Branch')
plt.yticks(fontsize=6,rotation=45)

plt.show()
```
